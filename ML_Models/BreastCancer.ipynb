```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/My Drive/datab.csv')
```

```python
df.head()
```

```python
df
```

```python
df = df.drop(columns='Patient_ID',axis=1)
df.head()
df = df.drop(columns='Age',axis=1)
df.head()
```

```python
df.head()
```

```python
df['Risk_Level']=df['Risk_Level'].map({'High':2,'Moderate':1,'Low':0})
df['Family_History'] = df['Family_History'].map({'No': 0, 'Yes': 1})
df['High_Fat_Diet'] = df['High_Fat_Diet'].map({'No': 0, 'Yes': 1})
df['Smoking'] = df['Smoking'].map({'No': 0, 'Yes': 1})
df['Alcohol'] = df['Alcohol'].map({'No': 0, 'Yes': 1})
df['HRT_Usage'] = df['HRT_Usage'].map({'No': 0, 'Yes': 1})
df['Late_Menopause'] = df['Late_Menopause'].map({'No': 0, 'Yes': 1})
df['No_or_Late_Pregnancy'] = df['No_or_Late_Pregnancy'].map({'No': 0, 'Yes': 1})
df['Breastfed'] = df['Breastfed'].map({'No': 0, 'Yes': 1})
df['Estrogen_Cosmetics'] = df['Estrogen_Cosmetics'].map({'No': 0, 'Yes': 1})
```

```python
df.head()
```

```python
corr = df.corr()
corr
```

```python
prediction_vars = ['Family_History', 'High_Fat_Diet', 'Smoking', 'Alcohol', 'HRT_Usage', 'Late_Menopause', 'No_or_Late_Pregnancy', 'Breastfed', 'Estrogen_Cosmetics']
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df,test_size=0.15,random_state=1)
```

```python
train_x = train[prediction_vars]
train_y = train['Risk_Level']
test_x = test[prediction_vars]
test_y = test['Risk_Level']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
model = MLPClassifier(hidden_layer_sizes=(100,100,100),max_iter=500)
```

```python
model.fit(train_x,train_y)
```

```python
predicted = model.predict(test_x)
```

```python
predicted = model.predict(test_x)

from sklearn.metrics import (
    confusion_matrix, precision_score, recall_score,
    accuracy_score, f1_score, classification_report
)

print("Confusion Matrix:")
print(confusion_matrix(test_y, predicted))

print("\nMetrics:")
print(f"Accuracy : {accuracy_score(test_y, predicted):.2f}")
print(f"Precision: {precision_score(test_y, predicted, average='weighted', zero_division=0):.2f}")
print(f"Recall   : {recall_score(test_y, predicted, average='weighted', zero_division=0):.2f}")
print(f"F1 Score : {f1_score(test_y, predicted, average='weighted', zero_division=0):.2f}")

print("\nClassification Report:")
print(classification_report(test_y, predicted, zero_division=0))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import joblib

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", RandomForestClassifier(
        n_estimators=300,
        class_weight="balanced",
        random_state=42
    ))
])

pipeline.fit(train_x, train_y)

joblib.dump(pipeline, "breast_risk_model.joblib")
```
